```python
import os
import re
from rouge import Rouge
data_path = "./prep_data/"
```

```python

reference = []
candidate = []

with open(os.path.join(data_path, 'valid.src'), encoding = 'utf-8') as f:
    for line in f.readlines():
        candidate.append(" ".join(re.split("\.|。|\?|\!|？|！|\:|：|,|，", line)[:3]))
with open(os.path.join(data_path, 'valid.tgt'), encoding = 'utf-8') as f:
    for line in f.readlines():
        reference.append(line)
```

```python
rouge = Rouge()
scores = rouge.get_scores(candidate, reference, avg = True)
scores
```
